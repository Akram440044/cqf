## Analytical Solution

$$ dS = rS \ dt + vS\ dW $$

Geometric Brownian Motion can be shown to have a solution,

$$ f(S;\mu, \sigma) = \frac{1}{S\sigma \sqrt{2\pi}} exp\big[ -\frac{ (logS - \mu)^2 }{2\sigma^2} \big] $$

The probability density function follows a lognormal distribution,

$$ f(S;\mu,\sigma) = PDF_{LogNormal} (S;\mu,\sigma) $$

$$ \mu = \bigg[ (r-\frac{1}{2}v^2)T + logS_0 \bigg] $$

$$\sigma = v\sqrt{T}$$

Finally, the expected value, or mean of the lognormal distribution can be computed as,

$$ \bar{S} = E[S] = exp\big( \mu + \frac{1}{2}\sigma^2 \big) $$

```python
# CODE 4.1 Lognormal distribution PDF graph
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: INPUTS 
mu, sigma = 1.0, 0.5

# STEP 2: PROCESS 
x = np.linspace(0.1, 10, 100)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
exact_mean = np.exp(mu + sigma**2/2)

# STEP 3: OUTPUTS 
plt.vlines(exact_mean, 0, 1.5*np.max(pdf), linestyle="dashed")
plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')
plt.show()

print('___________ parameters ___________')
print('mu = ',mu, ' sigma = ',sigma)
print(' ')
print('____________ results _____________')
print('exact_mean = ',exact_mean)
print('__________________________________')
```

## Numerical Solution

The numercial Soluton of the GBM can be obtained by discretising GBM as,

$$dS = rSdt + vS dW$$
$$\Delta S = rS\Delta t + vS \Delta W$$
$$S^{i+1} - S^i = rS^i\Delta t + vS^i \epsilon_i \sqrt{\Delta t}$$
$$S^{i+1} = S^i + rS^i\Delta t + vS^i \epsilon_i \sqrt{\Delta t}$$
$$S^{i+1} = S^i \bigg(1 + r\Delta t + v \epsilon_i \sqrt{\Delta t}\bigg)$$

```python
# CODE 4.2 Lognormal distribution MC sampling
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: inputs
mu, sigma = 1.0, 0.5

# STEP 2: samples from Lognormal distribution 
s = np.random.lognormal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 100, density=True, align='mid',alpha=0.5)

# STEP 3: generate Lognormal pdf
x = np.linspace(0.1, 10, 100)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))
exact_mean = np.exp(mu + sigma**2/2)

# STEP 4: generate plots
plt.vlines(exact_mean, 0, 1.5*np.max(pdf), linestyle="dashed")
plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')
plt.show()

print('___________ parameters ___________')
print('mu = ',mu, ' sigma = ',sigma)
print(' ')
print('____________ results _____________')
print('Exact_mean      = ',exact_mean)
print('MC_sample_mean  = ',np.mean(s))
print('__________________________________')
```

## Code 4.3 Numerical Solution

In this laboratory we ought to follow the trajectory of the underlying variable up to some future time T. Each path represents a possible scenario or evolution of the underlying price into the future. The code that we present here implements the iterated process described in the numerical solution section described before.

This is based in both random number generation and Monte Carlo simulation. The code is composed of three parts:

- The first Part, the necessary libraries are loaded and the parameters of the problem are specified.

- The second part represents the iterated process that is required to advance from time T = 0 to T. And the the M repetitions required. These are captured using a double for loop.

- The third and final part is where the average value is computed and the results presented in graphical form.

```python
# CODE 4.3 Geometric Brownian Motion with Monte Carlo 
import math
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: INPUTS
S0, T, vol, r = 1, 1, 0.30, 0.1 # GBM parameters
M = 1000 # MC simulations
N = 100 # MC timesteps
dt = T / N
S = np.zeros((N+1,M+1))
ST=np.zeros(M)

# STEP 2: Lognormal distribution
mu = ((r - 0.5 * vol**2) * T + np.log(S0))
sigma = vol * np.sqrt(T)
exact_mean = np.exp(mu + sigma**2/2)

# STEP 3: Generate trajectories using Geometric Brownian Motion
for j in range(M):
    S[0,j]=S0
    for i in range(N):
        epsilon =  sp.random.randn(1)
        S[i+1,j] = S[i,j]*(1+r*dt+vol*math.sqrt(dt)*epsilon)
    ST[j]=S[i+1,j]

# STEP 4: plot trajectories 
t = np.linspace(0,1,N+1)
plt.plot(t,S[:,1:50])
plt.axis('tight')
plt.show()

print('___________ parameters ___________')
print('S0=',S0,' vol=',vol,' r=',r,' T=',T)
print('N = ', N, ' M = ',M)
print('mu = ',mu, ' sigma = ',sigma)
print(' ')
print('____________ results _____________')
print('Exact_mean = ',exact_mean)
print('MC_ST_mean = ',np.mean(ST))
print('__________________________________')
```

## Code 4.5 Classical Numerical Solution: composition

To study the behavior the payoff on the predicted values of the underlying using this technique, we can explore this in the following Python code. In this code we generates as in the previous one distribution of potential values for the underlying, and subsequently for each of them, we apply the payoff H. Once we have obtained M predicted values of the payoff based on M predicted values of the underlying, we can then compute the expected payoff simply by taking an arithmetic average

```python
# CODE 4.5 COMPOSED WITH IDENTITY
import math
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

# Step  1: INPUTS
S0, T, vol, r = 2, 1, 0.40, 0.1 # GBM parameters
M = 1000 # MC simulations
N = 100 # MC timesteps
dt = T / N
S = np.zeros((N+1,M+1))
ST=np.zeros(M)
H=np.zeros(M)

# Step 2: H(y) function for payoff (target)   
def H(y):
    target = y # in this case is simple identity
    return target

# Step 3: Lognormal
mu = ((r - 0.5 * vol**2) * T + np.log(S0))
sigma = vol * np.sqrt(T)
exact_mean = np.exp(mu + sigma**2/2)

# Step 4: Simulate GBM trajectories
for j in range(M):
    S[0,j]=S0
    for i in range(N):
        epsilon =  sp.random.randn(1)
        S[i+1,j] = S[i,j]*(1+r*dt+vol*math.sqrt(dt)*epsilon)
    ST[j]=S[i+1,j]
mean_H = np.mean(H(ST))
    
# Step 5: plots

# input histogram F: uncertainty model (probability)
plt.hist(ST, 50, density=True, align='mid')

# target function (payoff) G:
x = np.linspace(0.1,4,50)
plt.plot(x, H(x), linewidth=2, color='r')
plt.axis('tight')
plt.axis('square')
plt.xlim(0, 4)
plt.ylim(0, 4)

# output histogram H: 
plt.hist(H(ST), 50, density=True, align='mid', orientation='horizontal')

plt.vlines(exact_mean, 0, 3, linestyle="dashed")
plt.hlines(mean_H, 0, 3, linestyle="dashed")
plt.xlabel('F: input distribution')
plt.ylabel('H: output distribution')
plt.show()

# Step 6: print results
print('___________ parameters ___________')
print('S0=',S0,' vol=',vol,' r=',r,' T=',T)
print('N = ', N, ' M = ',M)
print('mu = ',mu, ' sigma = ',sigma)
print(' ')
print('____________ results _____________')
print('Exact mean LOGNORMAL =',exact_mean)
print('MC estimate E[H(ST)] =',mean_H )
print('__________________________________')
```
