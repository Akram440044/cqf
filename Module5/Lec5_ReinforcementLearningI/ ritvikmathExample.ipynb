Assuming that a person went to a village to be a visit professor for 300 days. There are only 3 restaurants, which representatively give untility for that person are $N(10,5), N(8,4), N(5,25)$. That person do not know the true distribution of desirity for those restaurants, and thus have to visit those restaurant to mimic that distribution.

There are two types of strategies.

1. Exploration: randomly choose from those three restaurants.
2. Exploitation: choose $\epsilon$ times to visit those restaurants to form the distribution, and the pick the one with highest expectation. Finally, spend the rest $1-\epsilon$ to that highest-expected restaurant. 

- $\epsilon$ the percentage used to exploit.
- $\rho$, the **regret** is defined as the difference between the best possible expected utility and the choice's corresponding utility. $\rho$ is the lower the better.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
class Bandit():
    
    def __init__(self, T=300, epsilon=0.1):
        self.T = T
        self.epsilon = epsilon
        self.expectation = 0
        self.best = 10 * self.T
        self.rho = None
        self.choice = None

    def update(self):
        pass
    
    def restaurant(self, typee):
        if typee==0:
            mu = 10
            sigma = 5
            return np.random.normal(mu,sigma)
        elif typee ==1:
            mu = 8
            sigma = 4
            return np.random.normal(mu,sigma)
        else:
            mu = 5
            sigma = 25
            return np.random.normal(mu,sigma)

    
    def exploration(self, p=1):
        '''
        If p is an int, then randomly choose the number of p sample to be the exploration set.
        
        If p is a float, then choose p% of sample to be exploration set.
        '''
        if (type(p) == float) or (p==1):
            p = int(p*self.T)
            type_list = pd.Series(np.random.choice([0, 1, 2], p))
        elif type(p)==int:
            type_list = pd.Series(np.random.choice([0, 1, 2], p))
        # save the choice list
        self.choice = type_list
        # save the expecation value
        self.expectation = type_list.apply(self.restaurant).sum()
        # save the regret
        self.rho = self.best - self.expectation
        
    def exploitation(self):
        if self.epsilon == 0:
            self.exploration
            return None
        # epsilon % of data are used to exploit
        exploitation_set = int(self.T * self.epsilon)
        # run exploit
        choice_list = pd.DataFrame(np.random.choice([0, 1, 2], exploitation_set))
        choice_list['value'] = choice_list[0].apply(self.restaurant)
        # get the desired choice from the exploit process
        max_type = choice_list.groupby(0).mean().idxmax()
        # save the choice list / expectation / regret
        self.choice = choice_list[0].append( np.repeat(max_type, self.T * (1-self.epsilon)) )
        self.expectation = self.choice.apply(self.restaurant).sum()
        self.rho = self.best - self.expectation
```

```python
epsilon_list = np.arange(0,1,0.1)
N = 100

exploration=[]
exploitation=[]
for epsilon in epsilon_list:

    bandit = Bandit(epsilon=epsilon)

    a=[]
    b=[]
    for i in range(N):
        bandit.exploration()
        a.append(bandit.rho)
        bandit.exploitation()
        b.append(bandit.rho)

    exploration.append(np.array(a).mean())
    exploitation.append(np.array(b).mean())
    
presentation = pd.DataFrame({'Epsilon':epsilon_list,'Explore':exploration, 'Exploit':exploitation})
presentation
```

```python
plt.plot(presentation.Epsilon, presentation.Explore, label = 'Explore')
plt.plot(presentation.Epsilon, presentation.Exploit, label = 'Exploit')
plt.legend()
plt.ylabel('Regret')
plt.xlabel('Epsilon')
plt.title('Exploit Dominates All the Way')
plt.show()
```

From the figure we can find that Exploitation dominate Exploration all the way for all the epsilon choice. However, Exploit converges to Explore as Epsilon goes to 1 (The whole set is randomly chosen), and close to 0.
