```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('fivethirtyeight')
```

# Stock Analysis

$$X_t \sim O(\sqrt{dt})$$

$$B.M.: \quad dX_t = \phi \sqrt{dt}$$

- Euler Maruyamu: E-M:
$$ S_{t+1} = S_t \bigg(1+ \mu dt + \sigma \ dX_t \bigg)$$

- Exact:
$$S_{t+1} = S_t e^{(\mu - \frac{1}{2}\sigma^2)dt + \sigma \ dX_t}$$

```python
Stock = 100
mu = 0.05
sigma = 0.2

T = 1
days = 252
dt = T/days


N = 252
series = pd.DataFrame(np.zeros([N,2]), columns=['EM','Exact'], index = (np.zeros(N) + dt).cumsum())
series.loc[0] = Stock
series = series.sort_index()
```

```python
phi = np.random.randn(N)
dX = phi * np.sqrt(dt)


for i in range(N):
    series.iloc[i+1,0] = series.iloc[i,0] * (1 + mu * dt + sigma * dX[i]) # for EM
    series.iloc[i+1,1] = series.iloc[i,1] * np.exp( (mu - 1/2 * np.square(sigma)) * dt + sigma * dX[i] )    # for Exact
plt.plot(series)
plt.legend(['EM','Exact'])
plt.title('The EM price is coverd by the Exact')
plt.show()
```

# Interest Rate Analysis - Vasicek

$$Vasicek: \quad dr = (\eta - \gamma \ r)dt + \sigma \ dX$$

- $\gamma$ - gamma, the speed of reversion
- $\bar{r} = \frac{\eta}{\gamma}$ - the mean
- $ \eta = \gamma \bar{r} $

$$  r_{i+1} = r_i + (\eta - \gamma \ r_i)dt + \sigma \ dX_i  $$

```python
rate = 5.5 / 100 # i.e. today's rate 5.5%
eta = 0.08
gamma = 10 # 6 means half of a year; 52 means a week; 10 means 10 month
sigma = 0.015

T = 1
days = 252
dt = T/days

N = 252
series2 = pd.DataFrame(np.zeros([N,2]), columns=['IntRate','Exact'], index = (np.zeros(N) + dt).cumsum())
series2.loc[0] = rate
series2 = series2.sort_index()
```

```python
phi = np.random.randn(N)
dX = phi * np.sqrt(dt)

for i in range(N):
    series2.iloc[i+1,0] = series2.iloc[i,0]  + (eta - gamma * series2.iloc[i,0] )*dt + sigma * dX[i]
plt.plot(series2['IntRate'])
plt.title(f'Mean r-bar is eta/gamma = {eta/gamma*100}%')
plt.show()
```

Recall, the move `-r` in front.

$$dr = -r( \gamma -\bar{r})dt + \sigma \ dX$$

If we switch off the randomness, then

$$dr = -r( \gamma -\bar{r})dt$$

Integrate,

$$ r = \bar{r} + A\ e^{-\gamma t} $$

```python
rate = 5.5 / 100 # i.e. today's rate 5.5%
eta = 0.08
gamma = 10 # 6 means half of a year; 52 means a week; 10 means 10 month
sigma = 0

T = 1
days = 252
dt = T/days

N = 252
series2 = pd.DataFrame(np.zeros([N,2]), columns=['IntRate','Exact'], index = (np.zeros(N) + dt).cumsum())
series2.loc[0] = rate
series2 = series2.sort_index()

phi = np.random.randn(N)
dX = phi * np.sqrt(dt)

for i in range(N):
    series2.iloc[i+1,0] = series2.iloc[i,0]  + (eta - gamma * series2.iloc[i,0] )*dt + sigma * dX[i]
plt.plot(series2['IntRate'])
plt.title(f'Mean r-bar is eta/gamma = {eta/gamma*100}%')
plt.show()
```

# Simulation

```python
Stock = 100
mu = 0.2
sigma = 0.2

T = 1
days = 252
dt = T/days


N = 252


def generate_a_series(i):
    series = pd.DataFrame(np.zeros([N,1]), columns=[f'Stock_{i}'], index = (np.zeros(N) + dt).cumsum())
    series.loc[0] = Stock
    series = series.sort_index()
    phi = np.random.randn(N)
    dX = phi * np.sqrt(dt)
    for i in range(N):
        series.iloc[i+1,0] = series.iloc[i,0] * (1 + mu * dt + sigma * dX[i]) # for EM
    return series
```

```python
Num_series = 50

series_list = []
for i in range(Num_series):
    series_list.append( generate_a_series(i) )
    
series_list = pd.concat(series_list, axis = 1)

plt.figure(figsize=[10,6])
plt.plot(series_list)
plt.title(f'Number of Series Simulated is {Num_series}')
plt.show()
```
