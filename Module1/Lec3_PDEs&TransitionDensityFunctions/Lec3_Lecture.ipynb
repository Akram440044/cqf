# Trinomial random Walk

```python
a = 0.2
dt = 1
dy = 1
N = 5    # 5 lines
T = 100  # 100 movements each line


move = [-dy, 0, dy]
prob = [a, 1-2*a, a]

###################
list_each = []
list_all = []
walk = 0

for i in range(N):
    walk = 0
    list_each = [walk]
    for j in range(T):
        walk += np.random.choice(move, p = prob)
        list_each.append(walk)
    list_all.append(list_each)
    
nodes = np.vstack(list_all).T

plt.figure()
plt.plot(nodes)
plt.show()
```
