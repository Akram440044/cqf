```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

# Brownian Motion

```python
t = 1
n = 100

dt = t / n
dy = np.sqrt(dt) 
```

```python
def brownian_motion(n, t=1,):
    move = [0]
    dt = t / n
    dy = np.sqrt(dt) 
    for i in range(n):
        move.append(np.random.choice([-dy, dy])) 
    move = np.array(move)
    plt.plot( np.linspace(0,t,n+1) , move.cumsum())
    s_move = move.cumsum()
    quadratic_var = np.square(s_move[:-1] - s_move[1:]).sum()
    print(quadratic_var)
    
plt.figure(figsize=(10,8))
brownian_motion(n = 10, t = 1)
brownian_motion(n = 100, t = 1)
brownian_motion(n = 500, t = 1)
brownian_motion(n = 1000, t = 1)
print('The Quadratic Variation goes to one, as n goes to infinite.')
```
